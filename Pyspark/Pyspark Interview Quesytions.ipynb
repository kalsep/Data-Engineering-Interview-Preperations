# SparkContext vs SparkSession

## 🔹 SparkContext
- Core entry point (older API) in Apache Spark
- Used for RDD operations
- Handles cluster connection and resource management
- Only **one SparkContext per JVM**
- Low-level API

---

## 🔹 SparkSession
- Modern entry point (Spark 2.x+)
- Replaces SparkContext, SQLContext, HiveContext
- Used for DataFrame, Dataset, and SQL operations
- Internally contains SparkContext
- High-level API

---

## ⚡ Key Differences

| Feature        | SparkContext        | SparkSession              |
|----------------|--------------------|---------------------------|
| API Level      | Low-level          | High-level                |
| Introduced     | Early Spark        | Spark 2.0+                |
| Usage          | RDD                | DataFrame, SQL, Dataset   |
| Entry Point    | Yes (old)          | Yes (modern)              |
| Contains       | —                  | SparkContext              |

---

## 🔥 Relationship

SparkSession → SparkContext → Cluster

---

# RDD vs DataFrame vs Dataset

## 🔹 RDD (Resilient Distributed Dataset)
- Low-level abstraction in Apache Spark
- Distributed collection of objects
- No schema (unstructured data)
- Full control over transformations
- No built-in optimization

### ✅ Features:
- Immutable
- Type-safe (Scala/Java)
- Functional transformations (map, filter, reduce)
- Manual optimization required

---

## 🔹 DataFrame
- High-level abstraction (like a table)
- Has schema (columns + data types)
- Optimized using Catalyst Optimizer

### ✅ Features:
- SQL-like operations
- Better performance than RDD
- Easy to use
- Not type-safe (in Scala/Java)

---

## 🔹 Dataset
- Combines RDD + DataFrame
- Type-safe and optimized
- Available in Scala/Java

### ✅ Features:
- Compile-time type safety
- Uses Catalyst Optimizer
- Best of both worlds

---

## ⚡ Key Differences

| Feature        | RDD              | DataFrame        | Dataset          |
|----------------|------------------|------------------|------------------|
| Abstraction    | Low-level        | High-level       | High-level       |
| Schema         | ❌ No            | ✅ Yes           | ✅ Yes           |
| Performance    | ❌ Slower        | ✅ Faster        | ✅ Faster        |
| Type Safety    | ✅ Yes           | ❌ No            | ✅ Yes           |
| Optimization   | ❌ Manual        | ✅ Automatic     | ✅ Automatic     |
| Ease of Use    | ❌ Harder        | ✅ Easy          | ⚖️ Moderate      |

---

# 🎯 When to Choose RDD over DataFrame

Use RDD when:

## ✅ 1. Need Low-Level Control
- Custom partitioning
- Fine-grained transformations

## ✅ 2. Working with Unstructured Data
- No predefined schema (raw logs, text data)

## ✅ 3. Complex Transformations
- Logic not easily expressed using DataFrame/SQL APIs

## ✅ 4. Functional Programming Use Cases
- Heavy use of map, flatMap, reduce

## ✅ 5. Legacy Code
- Existing systems built on RDD

---

# 🚫 When NOT to Use RDD

Avoid RDD when:
- DataFrame/Dataset can solve the problem
- Performance is critical
- You want automatic optimization
- You prefer cleaner and maintainable code

---

## 💡 Rule of Thumb

- Use DataFrame → Default choice (most use cases)
- Use Dataset → When type safety is required (Scala/Java)
- Use RDD → Only for special low-level scenarios

# What Happens If Multiple SparkSessions Are Created?

## 🔹 Short Answer
- Multiple SparkSessions can be created
- They usually share the same SparkContext
- Only one SparkContext is allowed per JVM

---

## 🔹 Default Behavior (getOrCreate)

```python
spark1 = SparkSession.builder.getOrCreate()
spark2 = SparkSession.builder.getOrCreate()
````

### ✅ Result:

* Both variables refer to the same SparkSession
* No new session is created
* Existing session is reused

---

## 🔹 Creating a New Session (newSession)

```python
spark2 = spark1.newSession()
```

### ✅ Result:

* A new SparkSession is created
* Shares the same SparkContext
* Allows different configurations

---

## 🔹 Key Points

### 1. Single SparkContext

* Only one SparkContext can exist per JVM
* Attempting to create another → error

---

### 2. Shared Resources

* Same cluster
* Same executors
* Same underlying data/cache (mostly)

---

### 3. Independent Configurations

* Each session can have different SQL settings
* Useful for isolation

---

### 4. Temporary Views

* Session-scoped temporary views are isolated
* Global temporary views are shared

---

## 🔹 Internal Relationship

SparkSession → SparkContext → Cluster

---

## 🎯 Analogy

* SparkContext = Engine
* SparkSession = Driver

👉 One engine, multiple drivers

---

## 💡 Interview Answer

"Multiple SparkSessions can exist, but they share the same SparkContext. Only one SparkContext is allowed per JVM. New sessions can have separate SQL configurations but use the same underlying resources."

```


### 4. Transformations vs Actions
A key PySpark concept: Transformations are lazy, Actions trigger computation.
- Transformations: Operations like map(), filter(), select() that build a logical plan but don’t execute immediately.
- Actions: Operations like collect(), count(), show() that trigger execution and return results.

# Lazy Evaluation in Spark

## 🔹 Definition
Lazy evaluation means transformations are not executed immediately.
Execution happens only when an action is called.

---

## 🔹 Transformations vs Actions

### Transformations (Lazy)
- map()
- filter()
- select()

👉 Builds a logical plan (DAG), no execution

### Actions (Trigger Execution)
- collect()
- count()
- show()

👉 Triggers actual execution

---

## 🔹 Example

```python
rdd = sc.textFile("data.txt")
rdd2 = rdd.filter(lambda x: "error" in x)
rdd3 = rdd2.map(lambda x: x.upper())

rdd3.collect()  # Execution happens here

### Partitions & Parallelism
Partitioning determines how data is split across the cluster — it directly affects performance and parallelism.


Default partitions = number of cores
- repartition() → increases partitions (full shuffle)
- coalesce() → reduces partitions (no shuffle)

# repartition() vs coalesce()

## 🔹 repartition()
- Can increase or decrease partitions
- Performs full shuffle
- Ensures even data distribution

### ✅ Use Cases:
- Increasing partitions
- Fixing data skew
- Balancing data

---

## 🔹 coalesce()
- Mainly reduces partitions
- Avoids full shuffle
- Faster but may create imbalance

### ✅ Use Cases:
- Reducing partitions
- Writing fewer output files

---

## ⚡ Key Differences

| Feature            | repartition() | coalesce() |
|--------------------|--------------|-------------|
| Shuffle            | Yes          | No          |
| Performance        | Slower       | Faster      |
| Increase Partitions| Yes          | Not ideal   |
| Decrease Partitions| Yes          | Yes         |
| Distribution       | Even         | Uneven      |

---

# Partition Size and Performance

## ❌ Too Few Partitions
- Low parallelism
- Slow execution

## ❌ Too Many Partitions
- High overhead
- Too many small tasks

## ✅ Optimal
- 2–4 partitions per core

## ⚠ Data Skew
- Uneven partitions
- Slower tasks

## 📌 Impact
- Affects CPU usage, memory, and I/O
