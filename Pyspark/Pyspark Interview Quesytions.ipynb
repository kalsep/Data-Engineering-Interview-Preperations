# SparkContext vs SparkSession

## 🔹 SparkContext
- Core entry point (older API) in Apache Spark
- Used for RDD operations
- Handles cluster connection and resource management
- Only **one SparkContext per JVM**
- Low-level API

---

## 🔹 SparkSession
- Modern entry point (Spark 2.x+)
- Replaces SparkContext, SQLContext, HiveContext
- Used for DataFrame, Dataset, and SQL operations
- Internally contains SparkContext
- High-level API

---

## ⚡ Key Differences

| Feature        | SparkContext        | SparkSession              |
|----------------|--------------------|---------------------------|
| API Level      | Low-level          | High-level                |
| Introduced     | Early Spark        | Spark 2.0+                |
| Usage          | RDD                | DataFrame, SQL, Dataset   |
| Entry Point    | Yes (old)          | Yes (modern)              |
| Contains       | —                  | SparkContext              |

---

## 🔥 Relationship

SparkSession → SparkContext → Cluster

# RDD vs DataFrame vs Dataset

## 🔹 RDD (Resilient Distributed Dataset)
- Low-level abstraction in Apache Spark
- Distributed collection of objects
- No schema (unstructured data)
- Full control over transformations
- No built-in optimization

### ✅ Features:
- Immutable
- Type-safe (Scala/Java)
- Functional transformations (map, filter, reduce)
- Manual optimization required

---

## 🔹 DataFrame
- High-level abstraction (like a table)
- Has schema (columns + data types)
- Optimized using Catalyst Optimizer

### ✅ Features:
- SQL-like operations
- Better performance than RDD
- Easy to use
- Not type-safe (in Scala/Java)

---

## 🔹 Dataset
- Combines RDD + DataFrame
- Type-safe and optimized
- Available in Scala/Java

### ✅ Features:
- Compile-time type safety
- Uses Catalyst Optimizer
- Best of both worlds

---

## ⚡ Key Differences

| Feature        | RDD              | DataFrame        | Dataset          |
|----------------|------------------|------------------|------------------|
| Abstraction    | Low-level        | High-level       | High-level       |
| Schema         | ❌ No            | ✅ Yes           | ✅ Yes           |
| Performance    | ❌ Slower        | ✅ Faster        | ✅ Faster        |
| Type Safety    | ✅ Yes           | ❌ No            | ✅ Yes           |
| Optimization   | ❌ Manual        | ✅ Automatic     | ✅ Automatic     |
| Ease of Use    | ❌ Harder        | ✅ Easy          | ⚖️ Moderate      |

---

# 🎯 When to Choose RDD over DataFrame

Use RDD when:

## ✅ 1. Need Low-Level Control
- Custom partitioning
- Fine-grained transformations

## ✅ 2. Working with Unstructured Data
- No predefined schema (raw logs, text data)

## ✅ 3. Complex Transformations
- Logic not easily expressed using DataFrame/SQL APIs

## ✅ 4. Functional Programming Use Cases
- Heavy use of map, flatMap, reduce

## ✅ 5. Legacy Code
- Existing systems built on RDD

---

# 🚫 When NOT to Use RDD

Avoid RDD when:
- DataFrame/Dataset can solve the problem
- Performance is critical
- You want automatic optimization
- You prefer cleaner and maintainable code

---

## 💡 Rule of Thumb

- Use DataFrame → Default choice (most use cases)
- Use Dataset → When type safety is required (Scala/Java)
- Use RDD → Only for special low-level scenarios

````

### ✅ Result:

* Both variables refer to the same SparkSession
* No new session is created
* Existing session is reused

---

## 🔹 Creating a New Session (newSession)

```python
spark2 = spark1.newSession()
```

### ✅ Result:

* A new SparkSession is created
* Shares the same SparkContext
* Allows different configurations

---

## 🔹 Key Points

### 1. Single SparkContext

* Only one SparkContext can exist per JVM
* Attempting to create another → error

---

### 2. Shared Resources

* Same cluster
* Same executors
* Same underlying data/cache (mostly)

---

### 3. Independent Configurations

* Each session can have different SQL settings
* Useful for isolation

---

### 4. Temporary Views

* Session-scoped temporary views are isolated
* Global temporary views are shared

---

## 🔹 Internal Relationship

SparkSession → SparkContext → Cluster

---

## 🎯 Analogy

* SparkContext = Engine
* SparkSession = Driver

👉 One engine, multiple drivers

---

## 💡 Interview Answer

"Multiple SparkSessions can exist, but they share the same SparkContext. Only one SparkContext is allowed per JVM. New sessions can have separate SQL configurations but use the same underlying resources."

```


### 4. Transformations vs Actions
A key PySpark concept: Transformations are lazy, Actions trigger computation.
- Transformations: Operations like map(), filter(), select() that build a logical plan but don’t execute immediately.
- Actions: Operations like collect(), count(), show() that trigger execution and return results.

# Lazy Evaluation in Spark

## 🔹 Definition
Lazy evaluation means transformations are not executed immediately.
Execution happens only when an action is called.

---

## 🔹 Transformations vs Actions

### Transformations (Lazy)
- map()
- filter()
- select()

👉 Builds a logical plan (DAG), no execution

### Actions (Trigger Execution)
- collect()
- count()
- show()

👉 Triggers actual execution

---

## 🔹 Example

```python
rdd = sc.textFile("data.txt")
rdd2 = rdd.filter(lambda x: "error" in x)
rdd3 = rdd2.map(lambda x: x.upper())

rdd3.collect()  # Execution happens here

### Partitions & Parallelism
Partitioning determines how data is split across the cluster — it directly affects performance and parallelism.


Default partitions = number of cores
- repartition() → increases partitions (full shuffle)
- coalesce() → reduces partitions (no shuffle)

# repartition() vs coalesce()

## 🔹 repartition()
- Can increase or decrease partitions
- Performs full shuffle
- Ensures even data distribution

### ✅ Use Cases:
- Increasing partitions
- Fixing data skew
- Balancing data

---

## 🔹 coalesce()
- Mainly reduces partitions
- Avoids full shuffle
- Faster but may create imbalance

### ✅ Use Cases:
- Reducing partitions
- Writing fewer output files

---

## ⚡ Key Differences

| Feature            | repartition() | coalesce() |
|--------------------|--------------|-------------|
| Shuffle            | Yes          | No          |
| Performance        | Slower       | Faster      |
| Increase Partitions| Yes          | Not ideal   |
| Decrease Partitions| Yes          | Yes         |
| Distribution       | Even         | Uneven      |

---

# Partition Size and Performance

## ❌ Too Few Partitions
- Low parallelism
- Slow execution

## ❌ Too Many Partitions
- High overhead
- Too many small tasks

## ✅ Optimal
- 2–4 partitions per core

## ⚠ Data Skew
- Uneven partitions
- Slower tasks

## 📌 Impact
- Affects CPU usage, memory, and I/O

Handling schema evolution in PySpark is a core skill when working with evolving datasets—especially in data lakes, streaming pipelines, and ETL jobs. Let’s break it down clearly from fundamentals to advanced patterns.

---

# 🔁 What is Schema Evolution?

Schema evolution refers to how your data processing system handles **changes in data structure over time**, such as:

* Adding/removing columns
* Changing data types
* Renaming fields
* Nested structure changes

Example:

```json
// Old schema
{ "id": 1, "name": "Alice" }

// New schema
{ "id": 1, "name": "Alice", "age": 30 }
```

---

# 🧠 Why It Matters in PySpark

In PySpark (part of Apache Spark), schema evolution is important because:

* Data often comes from multiple sources with inconsistent formats
* Files (especially Parquet/JSON) may have different schemas
* Streaming pipelines must handle changing input safely

---

# 🛠️ Core Techniques to Handle Schema Evolution

## 1. Schema Merging (Parquet/ORC)

PySpark can automatically merge schemas across files.

### Example:

```python
df = spark.read.option("mergeSchema", "true").parquet("path/")
```

### When to use:

* Multiple Parquet files with different columns
* Data lake ingestion

### Pros:

✔ Automatically combines schemas
✔ Easy to use

### Cons:

❌ Slower (extra metadata scan)
❌ Not ideal for large datasets frequently

---

## 2. Explicit Schema Definition

Define schema manually instead of inferring it.

```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType

schema = StructType([
    StructField("id", IntegerType(), True),
    StructField("name", StringType(), True),
    StructField("age", IntegerType(), True)
])

df = spark.read.schema(schema).json("path/")
```

### Why:

* Ensures consistency
* Prevents unexpected failures

---

## 3. Handling Missing / Extra Columns

### Add missing columns:

```python
from pyspark.sql.functions import lit

df = df.withColumn("age", lit(None))
```

### Drop extra columns:

```python
df = df.select("id", "name")
```

---

## 4. Union with Different Schemas

Use:

```python
df1.unionByName(df2, allowMissingColumns=True)
```

### Why:

* Safely merges datasets with different columns
* Automatically fills missing columns with nulls

---

## 5. Schema Evolution in Delta Tables

If you’re using Delta Lake:

### Enable schema evolution:

```python
df.write.format("delta") \
  .option("mergeSchema", "true") \
  .mode("append") \
  .save("path/")
```

### Or auto-merge:

```python
spark.conf.set("spark.databricks.delta.schema.autoMerge.enabled", "true")
```

### Benefits:

✔ Handles column additions automatically
✔ Supports ACID transactions
✔ Great for production pipelines

---

## 6. Handling Data Type Changes

### Strategy:

* Cast explicitly:

```python
df = df.withColumn("age", df["age"].cast("int"))
```

### Important:

* Spark does NOT automatically handle incompatible type changes safely
* You must control casting logic

---

## 7. Nested Schema Evolution

For JSON or complex structs:

```python
from pyspark.sql.functions import col

df = df.withColumn("address_city", col("address.city"))
```

Or rebuild struct:

```python
from pyspark.sql.functions import struct

df = df.withColumn("address", struct("city", "zip"))
```

---

## 8. Streaming Schema Evolution

In Structured Streaming:

* Schema must be stable
* Evolution is tricky

### Approaches:

* Use schema registry (external system)
* Restart stream when schema changes
* Use permissive mode:

```python
spark.readStream.option("mode", "PERMISSIVE")
```

---

## 9. Backward & Forward Compatibility

### Backward compatible:

* Adding new columns (safe)

### Forward compatible:

* Dropping columns carefully

### Breaking changes:

* Renaming columns
* Changing types

---

# 🧩 Best Practices

### ✅ DO:

* Define schemas explicitly in production
* Use `unionByName` instead of `union`
* Track schema versions
* Use Delta Lake for evolving pipelines

### ❌ AVOID:

* Relying on schema inference in production
* Frequent schema merging on large datasets
* Silent type changes

---

# 🧪 Real-World Pattern (Production)

Typical pipeline:

1. Ingest raw data (flexible schema)
2. Normalize schema (apply standard structure)
3. Store in Delta with evolution enabled
4. Enforce schema at consumption layer

---

# ⚡ Interview-Ready Summary

If asked in an interview:

> "In PySpark, I handle schema evolution using schema merging for file formats like Parquet, explicit schema definitions for consistency, and `unionByName` for combining datasets. In production, I prefer Delta Lake with auto schema evolution enabled. I also manage type changes explicitly and ensure backward compatibility while versioning schemas."

---

If you want, I can walk you through a **real-world ETL pipeline example with schema evolution step-by-step** or give **interview questions + answers on this topic**.


Here’s a clear, structured explanation of both concepts in **Apache Spark** 👇

---

# 📌 Schema Inference vs Schema Merging in Spark

## 🔍 1. Schema Inference

### 🧠 What is Schema Inference?

Schema inference is when Spark **automatically detects the structure (schema)** of your data instead of you defining it manually.

---

### ⚙️ How it works

When you read data (like JSON/CSV/Parquet), Spark scans the dataset and determines:

* Column names
* Data types (int, string, double, etc.)
* Nullable fields

---

### 💡 Example

```python
df = spark.read.json("data.json")
df.printSchema()
```

Spark might infer:

```
root
 |-- id: long
 |-- name: string
 |-- age: long
```

---

### ✅ Advantages

* Quick and convenient
* Useful for exploration and prototyping
* No need to define schema manually

---

### ❌ Disadvantages

* Slower (needs to scan data)
* Can infer wrong types (e.g., string instead of int)
* Inconsistent across files
* Not recommended for production pipelines

---

### 📌 When to Use

* Ad-hoc analysis
* Small datasets
* Initial data exploration

---

## 🔗 2. Schema Merging

### 🧠 What is Schema Merging?

Schema merging is when Spark **combines multiple schemas into one unified schema** when reading data from multiple files.

---

### ⚙️ How it works

If different files have different columns, Spark merges them:

* Includes all columns from all files
* Missing values are filled with `null`

---

### 💡 Example

**File 1:**

```json
{ "id": 1, "name": "Alice" }
```

**File 2:**

```json
{ "id": 2, "age": 30 }
```

---

### Code:

```python
df = spark.read.option("mergeSchema", "true").json("path/")
df.printSchema()
```

---

### Resulting Schema:

```
root
 |-- id: long
 |-- name: string
 |-- age: long
```

---

### ✅ Advantages

* Handles evolving data automatically
* Useful in data lakes with incremental data
* No need to preprocess files

---

### ❌ Disadvantages

* Performance overhead (metadata scan)
* Can become expensive on large datasets
* May introduce unexpected nulls

---

### 📌 When to Use

* Data lakes (Parquet/ORC)
* Slowly evolving schemas
* Batch ingestion pipelines

---

# ⚖️ Key Differences

| Feature        | Schema Inference   | Schema Merging          |
| -------------- | ------------------ | ----------------------- |
| Purpose        | Detect schema      | Combine schemas         |
| Input          | Single dataset     | Multiple files/datasets |
| Performance    | Slower (data scan) | Slower (metadata merge) |
| Control        | Low                | Medium                  |
| Production Use | ❌ Avoid            | ⚠️ Use carefully        |

---

# 🧠 Simple Analogy

* **Schema Inference** = “Let Spark guess the structure”
* **Schema Merging** = “Combine different structures into one”

---

# ⚡ Interview-Ready Answer

> Schema inference in Spark is the automatic detection of a dataset’s structure when reading data, while schema merging is the ability to combine schemas from multiple files into a unified schema. In production, schema inference is usually avoided due to inconsistency, while schema merging is used selectively for evolving datasets, especially with formats like Parquet.

---

If you want, I can give you **real interview questions**, **tricky edge cases**, or a **hands-on PySpark scenario** to practice these concepts.


Let’s break this down clearly—from concept to how PySpark actually deals with it in real systems.

---

# 📌 What is Data Skewness?

## 🧠 Definition

**Data skewness** occurs when data is **unevenly distributed across partitions**, meaning some partitions have significantly more data than others.

---

## 💡 Example

Imagine a dataset:

| user_id | country |
| ------- | ------- |
| 1       | India   |
| 2       | India   |
| 3       | India   |
| 4       | USA     |

If Spark partitions by `country`:

* One partition (India) → **huge data**
* Another (USA) → **very small**

👉 This imbalance = **data skew**

---

## ⚠️ Why It’s a Problem in Apache Spark

Spark works in parallel, so:

* Each partition → one task
* If one partition is huge → one task runs much longer

### Result:

* Slow jobs
* Resource underutilization
* Stragglers (one slow task delays entire job)

---

# 🔥 Where Data Skew Happens Most

* Joins (especially on skewed keys)
* GroupBy / aggregations
* Window functions
* Wide transformations

---

# 🛠️ How PySpark Handles Data Skew

## 1. 🔀 Adaptive Query Execution (AQE)

Spark (3.x+) automatically detects skew and splits large partitions.

```python
spark.conf.set("spark.sql.adaptive.enabled", "true")
spark.conf.set("spark.sql.adaptive.skewJoin.enabled", "true")
```

### What it does:

* Detects skewed partitions
* Splits them into smaller chunks
* Improves parallelism

---

## 2. 🧂 Salting Technique (Manual Fix)

Add randomness to skewed keys.

```python
from pyspark.sql.functions import rand, concat

df = df.withColumn("salted_key", concat(df["key"], (rand()*10).cast("int")))
```

### Idea:

* Break one heavy key into multiple smaller ones
* Distribute load evenly

---

## 3. 📦 Broadcast Join

If one dataset is small:

```python
from pyspark.sql.functions import broadcast

df.join(broadcast(small_df), "key")
```

### Why:

* Avoids shuffle entirely
* Prevents skew impact

---

## 4. 🔁 Repartitioning

```python
df = df.repartition(100, "key")
```

### Helps:

* Redistribute data more evenly

⚠️ But:

* Doesn’t fix skew if key itself is skewed

---

## 5. 🎯 Skew Join Hints

```python
df.join(other_df.hint("skew"), "key")
```

### Use when:

* You know skew exists
* Want Spark to optimize join

---

## 6. 🧮 Pre-Aggregation

Reduce skew before join:

```python
df.groupBy("key").count()
```

### Idea:

* Shrink large datasets early

---

## 7. 🧱 Handling Skewed Keys Separately

Split dataset:

```python
skewed = df.filter(df.key == "hot_key")
normal = df.filter(df.key != "hot_key")
```

Process separately, then combine.

---

# ⚖️ Summary of Techniques

| Technique       | When to Use                 | Automatic  |
| --------------- | --------------------------- | ---------- |
| AQE             | Spark 3+ jobs               | ✅ Yes      |
| Salting         | Heavy skewed keys           | ❌ No       |
| Broadcast Join  | Small lookup tables         | ⚠️ Partial |
| Repartition     | General distribution issues | ❌ No       |
| Skew Join Hint  | Known skew in joins         | ❌ No       |
| Pre-Aggregation | Before heavy joins          | ❌ No       |

---

# 🧠 Simple Analogy

* Data skew = **one cashier handling 90% of customers**
* Fix = open more counters or split the queue

---

# ⚡ Interview-Ready Answer

> Data skewness occurs when data is unevenly distributed across partitions, causing some tasks to take significantly longer than others. In PySpark, it is handled using techniques like Adaptive Query Execution (AQE) for automatic skew handling, salting to distribute skewed keys, broadcast joins to avoid shuffles, and repartitioning or pre-aggregation to balance data distribution.

---

If you want, I can give you:

* A **real skewed join example with before/after performance**
* **Spark UI screenshots explanation**
* **Advanced interview scenarios (FAANG-level)**


Let’s break down both concepts clearly—they’re tightly connected in real-time data pipelines built with **Apache Spark**.

---

# 📡 1. What is Structured Streaming?

## 🧠 Definition

**Structured Streaming** is a high-level API in PySpark that allows you to **process real-time data streams using the same DataFrame/Dataset API as batch processing**.

👉 Think of it as:

> “Treating streaming data like an unbounded table”

---

## 💡 How It Works

* Data continuously arrives (Kafka, files, sockets, etc.)
* Spark processes it in **micro-batches** (or continuous mode)
* You write queries just like batch DataFrames

---

## 🔄 Example

```python
df = spark.readStream.format("json").load("input/")
```

```python
query = df.writeStream \
    .format("console") \
    .start()
```

---

## ⚙️ Key Features

### ✅ Unified API

* Same code for batch & streaming

### ✅ Fault Tolerance

* Recovers from failures

### ✅ Exactly-Once Processing (with support systems)

### ✅ Scalability

* Distributed processing

---

## 📊 Common Use Cases

* Real-time analytics
* Log processing
* Fraud detection
* IoT data pipelines

---

# 💾 2. What is Checkpointing?

## 🧠 Definition

**Checkpointing** is the mechanism Spark uses to **store the progress and state of a streaming job** so it can recover from failures.

---

## 💡 Why It’s Needed

Streaming jobs:

* Run continuously
* Can fail (crash, restart, etc.)

👉 Without checkpointing:

* You lose progress
* Data may be reprocessed or lost

---

## 🔄 Example

```python
query = df.writeStream \
    .format("parquet") \
    .option("checkpointLocation", "chk/") \
    .start("output/")
```

---

## 📦 What Gets Stored in Checkpoint?

* Offsets (how much data has been processed)
* Metadata about the query
* Intermediate state (for aggregations)

---

## 🧩 Types of Checkpointing

### 1. Metadata Checkpointing

* Tracks progress (offsets, batch IDs)

### 2. State Checkpointing

* Stores aggregation/window state

---

# 🔗 Relationship Between Structured Streaming & Checkpointing

| Feature           | Structured Streaming     | Checkpointing  |
| ----------------- | ------------------------ | -------------- |
| Purpose           | Process real-time data   | Track progress |
| Runs continuously | ✅ Yes                    | ✅ Supports it  |
| Fault tolerance   | Depends on checkpointing | Core mechanism |

👉 Checkpointing is what makes Structured Streaming **fault-tolerant**

---

# ⚠️ Important Rules

* Each streaming query must have a **unique checkpoint location**
* Don’t delete checkpoint folder during execution
* Keep checkpoint in reliable storage (HDFS, S3, etc.)

---

# 🧠 Simple Analogy

* **Structured Streaming** = Live video stream 🎥
* **Checkpointing** = Resume playback point ⏯️

---

# ⚡ Interview-Ready Answer

> Structured Streaming in PySpark is a scalable and fault-tolerant stream processing engine that allows real-time data processing using the DataFrame API. Checkpointing is used within structured streaming to store progress and state information, enabling fault tolerance and exactly-once processing by allowing the system to recover from failures without data loss.

---

If you want, I can go deeper into:

* **Watermarking (very commonly asked with streaming)**
* **Exactly-once vs at-least-once semantics**
* **Kafka + PySpark real-time pipeline example**
